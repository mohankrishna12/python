# Functions
A zero or more inputs and returning a corresponding output


```python
def triple(x):
    """
    This functions multiplies the input by 3
    """
    return x*3
```

### Lambda Functions
A short anonymous functions or *lambdas*

```python
another_triple=lambda x:x*3
y=another_triple(4)    #12
```

Function parameters can also be given default arguments. If you specify some message while calling then it will be printed and if you won't specify any message inside argument body, it will print the default message as shown in below example.

```python
def my_message(message="Welcome to the simplified python repository"):
    print(message)
    
my_message()    #Welcome to the simplified python repository
my_message("Hello World!")    #Hello World!
```

It is useful to specify arguments by name

```python
def full_name(first_name="S", last_name="Py"):
    return print(first_name + " " + last_name)

full_name()    #S Py
full_name("Mohan", "Krishna")    #Mohan Krishna
full_name("Mohan", last_name="Krishna")    #Mohan Krishna
full_name(first_name="Mohan", last_name="Krishna")    #Mohan Krishna
full_name(last_name="Krishna", first_name="Mohan")    #Mohan Krishna
```

# Strings
Strings can be delimited by single or double quotation marks (but the quotes have to match):

```python
single_quote_string = 'Hello World!'
double_quote_string = "Hello World!"
tab_string = "\t" # represents a tab space 
print(tab_string)    #    (prints a tab space)

multi_line_string = """Hello!
Welcome to the simplified python repository"""   #multi-line string

print(multi_line_string)
```

If we have two names given separately, to combine them: there are multiple ways to do that, which are shown below. 

```python
first_name = "Mohan"
last_name = "Krishna"
full_name1 = first_name + " " + last_name
print(full_name1)    #Mohan Krishna
full_name2 = "{} {}".format(first_name, last_name)
print(full_name2)    #Mohan Krishna
full_name3 = f"{first_name} {last_name}"    #f-string  => preferred one
print(full_name3)    #Mohan Krishna
```

# Exceptions
Its always better to handle the exceptions, because exceptions cause your program to crash sometimes. By using *try* and *except*, lets handle with primitive example. 

```python
try:
    print(0/0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```
